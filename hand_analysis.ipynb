```python
import os
import matplotlib.pyplot as plt
```

```python
PATH = 'Desktop/fine/'
r_path = PATH+'rock'
s_path = PATH+'scissors'
p_path = PATH+'paper'
r_files = next(os.walk(r_path))[2]
s_files = next(os.walk(s_path))[2]
p_files = next(os.walk(p_path))[2]
```

```python
import pandas as pd
whole_list = list()
for idx, files in enumerate([r_files, s_files, p_files]):
    for file in files:
        pcas = file[0:-4].split('_')
        pc_array = list()
        pc_array.append(idx)

        for pc in pcas:
            pc_array.append(float(pc[1:-1]))
        whole_list.append(pc_array)
```

```python
cols = list()
cols.append('kind')
cols+=[f'pc{i}' for i in range(1,8)]
df = pd.DataFrame(whole_list, columns=cols)
df
```

```python
import seaborn as sns
for pc in cols[1:]:
#     ax = df.plot.scatter(x=pc, y='kind', hue='kind')
    ax = sns.scatterplot(data=df, x=pc, y='kind', hue='kind', palette=["C0", "C1", "k"])
    ax.set_title(pc)
    plt.show()
```
